```python
import random
from tkinter import *
from tkinter import messagebox

version = "0.3"

class Minesweeper(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        self.areas = None
        self.pinned = None
        self.count = None
        self.start()
        
        
        
    def start(self):#start variables
        self.widgets()
        
        self.areas = ["a", "b", "c","d",
                      "e", "f", "g", "h",
                      "i", "j", "k", "l",
                      "m", "n", "o",  "p"]
        
        self.areas2 = random.sample(self.areas, len(self.areas))
        
        self.bombs = []
        
        self.bomb_amount = random.randint(2, 7)
        
        self.position_bombs(self.areas2, self.bombs) #randomly chosen bomb areas
        
        self.pinned = [] #selected
        
        self.pinned2 = []
        
        self.value = 0
        
        minesweeper.title(f"Minesweeper {version}                             Mines: {len(self.bombs)}")
        
        #lines:
        self.L = ([self.areas2[0],self.areas2[1],self.areas2[2],self.areas2[3]], 
                  [self.areas2[4],self.areas2[5],self.areas2[6],self.areas2[7]],               
                  [self.areas2[8],self.areas2[9],self.areas2[10],self.areas2[11]], 
                  [self.areas2[12],self.areas2[13],self.areas2[14],self.areas2[15]],
                  ["z","z","z","z"])
        
        #columns:
        self.C = ([self.areas2[0],self.areas2[4],self.areas2[8],self.areas2[12]], 
                  [self.areas2[1],self.areas2[5],self.areas2[9],self.areas2[13]], 
                  [self.areas2[2],self.areas2[6],self.areas2[10],self.areas2[14]], 
                  [self.areas2[3],self.areas2[7],self.areas2[11],self.areas2[15]],
                  ["z","z","z","z"])
        
        
    def engine(self, x):
        if x not in self.pinned:
            if x in self.bombs:
                self.show_bombs(x)
                messagebox.showinfo("info", "you lose!")
                self.start()
            else:
                self.pinned.append(x)
                self.change_area(x) 
                self.values(x)
                if x in self.pinned2:
                    self.show_values(x)
                if len(self.pinned) >= (len(self.areas2) - len(self.bombs)):
                    messagebox.showinfo("congrats", "you win!")
                    self.start()
                
        else:
            messagebox.showinfo("info", "invalid value!")
            
    def values(self, x):#find bombs position
        self.value = 0
        self.pinned2.clear()
        n = self.areas2[self.get_index(self.areas2, x)]
        
        for i in self.bombs:
            B = self.areas2[self.get_index(self.areas2, i)]            
            if (n in self.L[self.get_grid(self.L, B)] or n in self.L[self.get_grid(self.L, B) -1] or n in self.L[self.get_grid(self.L, B) +1]) and\
            (n in self.C[self.get_grid(self.C, B)] or n in self.C[self.get_grid(self.C, B) -1] or n in self.C[self.get_grid(self.C, B) +1]):
                self.value +=1
                if len(self.pinned2) == 0:
                    self.pinned2.append(n)                        

    def get_index(self, itens, n):
        for i in itens:
            if n not in itens:
                return -1
            else:
                return itens.index(n)
    
    def get_grid(self, itens, n):
        for i in itens:
            for _ in i:
                if n not in i:
                    pass
                else:
                    return itens.index(i)
                
    def position_bombs(self, items, pinned):
        while True:
            n = random.choice(items)
            if n not in pinned:
                pinned.append(n)
                if len(pinned) == self.bomb_amount:
                    break
    
    def show_values(self, x):
        if self.value > 0:
            if x == self.areas2[0]:
                self.button0.configure(text=self.value)
            if x == self.areas2[1]:
                self.button1.configure(text=self.value)
            if x == self.areas2[2]:
                self.button2.configure(text=self.value)
            if x == self.areas2[3]:
                self.button3.configure(text=self.value)
            if x == self.areas2[4]:
                self.button4.configure(text=self.value)
            if x == self.areas2[5]:
                self.button5.configure(text=self.value)
            if x == self.areas2[6]:
                self.button6.configure(text=self.value)
            if x == self.areas2[7]:
                self.button7.configure(text=self.value)
            if x == self.areas2[8]:
                self.button8.configure(text=self.value)
            if x == self.areas2[9]:
                self.button9.configure(text=self.value)
            if x == self.areas2[10]:
                self.button10.configure(text=self.value)
            if x == self.areas2[11]:
                self.button11.configure(text=self.value)
            if x == self.areas2[12]:
                self.button12.configure(text=self.value)
            if x == self.areas2[13]:
                self.button13.configure(text=self.value)
            if x == self.areas2[14]:
                self.button14.configure(text=self.value)
            if x == self.areas2[15]:
                self.button15.configure(text=self.value)
        
    def show_bombs(self, x):
        if self.areas2[0] in self.bombs:
            self.button0.configure(bg="red")
        if self.areas2[1] in self.bombs:
            self.button1.configure(bg="red")
        if self.areas2[2] in self.bombs:
            self.button2.configure(bg="red")
        if self.areas2[3] in self.bombs:
            self.button3.configure(bg="red")
        if self.areas2[4] in self.bombs:
            self.button4.configure(bg="red")        
        if self.areas2[5] in self.bombs:
            self.button5.configure(bg="red")
        if self.areas2[6] in self.bombs:
            self.button6.configure(bg="red")
        if self.areas2[7] in self.bombs:
            self.button7.configure(bg="red")
        if self.areas2[8] in self.bombs:
            self.button8.configure(bg="red")
        if self.areas2[9] in self.bombs:
            self.button9.configure(bg="red")
        if self.areas2[10] in self.bombs:
            self.button10.configure(bg="red")
        if self.areas2[11] in self.bombs:
            self.button11.configure(bg="red")
        if self.areas2[12] in self.bombs:
            self.button12.configure(bg="red")
        if self.areas2[13] in self.bombs:
            self.button13.configure(bg="red")
        if self.areas2[14] in self.bombs:
            self.button14.configure(bg="red")
        if self.areas2[15] in self.bombs:
            self.button15.configure(bg="red")
            
    def change_area(self, x):
        if x == self.areas2[0]:
            self.button0.configure(bg="white")
        if x == self.areas2[1]:
            self.button1.configure(bg="white")
        if x == self.areas2[2]:
            self.button2.configure(bg="white")
        if x == self.areas2[3]:
            self.button3.configure(bg="white")
        if x == self.areas2[4]:
            self.button4.configure(bg="white")        
        if x == self.areas2[5]:
            self.button5.configure(bg="white")
        if x == self.areas2[6]:
            self.button6.configure(bg="white")
        if x == self.areas2[7]:
            self.button7.configure(bg="white")
        if x == self.areas2[8]:
            self.button8.configure(bg="white")
        if x == self.areas2[9]:
            self.button9.configure(bg="white")
        if x == self.areas2[10]:
            self.button10.configure(bg="white")
        if x == self.areas2[11]:
            self.button11.configure(bg="white")
        if x == self.areas2[12]:
            self.button12.configure(bg="white")
        if x == self.areas2[13]:
            self.button13.configure(bg="white")
        if x == self.areas2[14]:
            self.button14.configure(bg="white")
        if x == self.areas2[15]:
            self.button15.configure(bg="white")
            
    def widgets(self):        
        #show game
        self.frame1 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame1.grid(row=0)

        #buttons
        self.button0 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[0]))
        self.button0.grid(row=0, column=0)
        
        self.button1 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[1]))
        self.button1.grid(row=0, column=1)
        
        self.button2 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[2]))
        self.button2.grid(row=0, column=2)
        
        self.button3 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[3]))
        self.button3.grid(row=0, column=3)
        
        self.button4 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[4]))
        self.button4.grid(row=1, column=0)
        
        self.button5 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[5]))
        self.button5.grid(row=1, column=1)
        
        self.button6 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[6]))
        self.button6.grid(row=1, column=2)
        
        self.button7 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[7]))
        self.button7.grid(row=1, column=3)
        
        self.button8 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[8]))
        self.button8.grid(row=2, column=0)
        
        self.button9 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[9]))
        self.button9.grid(row=2, column=1)
        
        self.button10 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[10]))
        self.button10.grid(row=2, column=2)
        
        self.button11 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[11]))
        self.button11.grid(row=2, column=3)
        
        self.button12 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[12]))
        self.button12.grid(row=3, column=0)
        
        self.button13 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[13]))
        self.button13.grid(row=3, column=1)
        
        self.button14 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[14]))
        self.button14.grid(row=3, column=2)
        
        self.button15 = Button(self.frame1, font=("System", 12), text="   ", padx=35, pady=25, bg="#00008b", command=lambda: self.engine(self.areas2[15]))
        self.button15.grid(row=3, column=3)
               
minesweeper = Tk()
#minesweeper.title(f"Minesweeper {version}")
minesweeper.resizable(0,0)
app = Minesweeper(minesweeper).grid()
minesweeper.mainloop()       
```
